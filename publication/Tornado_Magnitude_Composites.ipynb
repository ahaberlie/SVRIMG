### Composites organized by EF-scale

```python
import matplotlib.pyplot as plt
from map_helper import radar_colormap
from matplotlib.colors import BoundaryNorm
import numpy as np
import pandas as pd
%matplotlib inline

plt.rcParams['figure.figsize'] = 20, 20
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 25
plt.rcParams['axes.labelsize'] = 25

cmap = radar_colormap()
classes = np.array(list(range(0, 85, 5)))
norm = BoundaryNorm(classes, ncolors=cmap.N)

idx = [1, 2, 4, 5, 7, 8]

for i in range(0, 6):
    
    img = np.load("../data/mag/tor/{}_all_tor_mag_pmm.npy".format(i))
    df_ = pd.read_csv("../data/mag/tor/{}_all_tor_mag_stats.csv".format(i))
    
    ax = plt.subplot(4,3,idx[i])
    ax.arrow(200, 190, 0, -0.5, head_width=20, head_length=30, fc='k', ec='k', zorder=10)
    ax.text(188, 215, "N", fontsize=30, zorder=10)
    mmp = ax.imshow(5*(img)+0.001, cmap=cmap, norm=norm)

    plt.colorbar(mmp, ax=ax, shrink=0.7, pad=0.01)

    ax.set_yticks(list(range(0, 300, 50)))
    ax.set_yticklabels([0, 100, 200, 300, 400, 500])

    ax.set_xticks(list(range(0, 300, 50)))
    ax.set_xticklabels([0, 100, 200, 300, 400, 500])

    ax.set_xlabel("km")
    ax.set_ylabel("km")

    ax.set_title('EF{}'.format(i), fontsize=30)
    ax.text(120, 245, "n={}".format(len(df_)), fontsize=30)
    ax.grid()
plt.tight_layout()
```
